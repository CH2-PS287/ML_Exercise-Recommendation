```python
# Create function to upload & download file in google colab, saved at content folder
import os
from google.colab import files

def uploadFile():
    uploaded = files.upload()
    for filename, content in uploaded.items():
        with open('/content/' + filename, 'wb') as f:
            f.write(content)

    os.listdir('/content')

def download_file(file_path):
    files.download(file_path)
```

```python
import random

def generate_random_value(base_value, deviation_range, total_decimal_places):
    """
    Generate a random value around the given base value with a specified deviation range.

    Parameters:
    - base_value (float): The base value around which the random value will be generated.
    - deviation_range (float): The range of deviation from the base value.
    - total_decimal_places (int): The total number of decimal places in the generated value.

    Returns:
    float: The randomly generated value.
    """
    random_deviation = random.uniform(-deviation_range, deviation_range)
    random_value = base_value + random_deviation
    rounded_random_value = round(random_value, total_decimal_places)

    return rounded_random_value
```

```python
def convert_to_comma(data):
    if isinstance(data, float):
        data = str(data)
    elif isinstance(data, int):
        data = str(data)
    return data.replace('.', ',')
```

```python
additional_data = 10
new_data = []
deviation_range = 1

for index, row in data.iterrows():
    new_data.append(pd.Series({
        'Activity, Exercise or Sport (1 hour)': row['Activity, Exercise or Sport (1 hour)'],
        '130 lb': convert_to_comma(row['130 lb']),
        '155 lb': convert_to_comma(row['155 lb']),
        '180 lb': convert_to_comma(row['180 lb']),
        '205 lb': convert_to_comma(row['205 lb']),
        'Calories per kg': convert_to_comma(row['Calories per kg']),
    }))
    for i in range(1, additional_data):
        new_data.append(pd.Series({
            'Activity, Exercise or Sport (1 hour)': row['Activity, Exercise or Sport (1 hour)'],
            '130 lb': convert_to_comma(generate_random_value(row['130 lb'], deviation_range, 2)),
            '155 lb': convert_to_comma(generate_random_value(row['155 lb'], deviation_range, 2)),
            '180 lb': convert_to_comma(generate_random_value(row['180 lb'], deviation_range, 2)),
            '205 lb': convert_to_comma(generate_random_value(row['205 lb'], deviation_range, 2)),
            'Calories per kg': convert_to_comma(generate_random_value(row['Calories per kg'], deviation_range, 14))
        }))
augmented_data = pd.DataFrame(new_data)
augmented_data
```

```python
augmented_data.to_csv('/content/exercise_dataset_augmented.csv', index=False)
download_file('/content/exercise_dataset_augmented.csv')
```
